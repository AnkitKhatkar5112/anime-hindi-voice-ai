# 🎌 Stage 1: Audio Exploration
Load, visualize, and understand the source audio before running the pipeline.

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import IPython.display as ipd
import numpy as np

AUDIO_PATH = '../data/processed/audio.wav'
audio, sr = librosa.load(AUDIO_PATH, sr=22050)
print(f'Duration: {len(audio)/sr:.2f}s | Sample Rate: {sr} Hz')
ipd.Audio(audio, rate=sr)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(14, 9))

# Waveform
librosa.display.waveshow(audio, sr=sr, ax=axes[0])
axes[0].set_title('Waveform')

# Spectrogram
D = librosa.stft(audio)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
img = librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='hz', ax=axes[1])
axes[1].set_title('Spectrogram')
plt.colorbar(img, ax=axes[1], format='%+2.0f dB')

# Mel spectrogram
M = librosa.feature.melspectrogram(y=audio, sr=sr)
M_db = librosa.power_to_db(M, ref=np.max)
img2 = librosa.display.specshow(M_db, sr=sr, x_axis='time', y_axis='mel', ax=axes[2])
axes[2].set_title('Mel Spectrogram')
plt.colorbar(img2, ax=axes[2], format='%+2.0f dB')

plt.tight_layout()
plt.show()
```

```python
# Pitch (F0) extraction — useful for prosody analysis
f0, voiced_flag, voiced_probs = librosa.pyin(
    audio, fmin=librosa.note_to_hz('C2'), fmax=librosa.note_to_hz('C7')
)
times = librosa.times_like(f0, sr=sr)

plt.figure(figsize=(14, 3))
plt.plot(times, f0, label='F0 (Pitch)', color='cyan', linewidth=1)
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('Pitch Contour (F0)')
plt.legend()
plt.tight_layout()
plt.show()
```
